```python
data = pd.read_csv("data_out2.csv")
```

```python
data.head()
```

```python
df = data[(data['sido']!='서울')&(data['sido']!='세종') &  (data['sido']!='울산') &  (data['sido']!='대구')&  (data['sido']!='부산')&  (data['sido']!='광주')&  (data['sido']!='인천') &  (data['sido']!='대전')]
```

```python
df['GB3'].unique()
```

```python
pv = pd.pivot_table(df, values='VLM', index=['sido','SGNG_NM'], columns='GB3', 
               aggfunc=np.mean, fill_value=0).reset_index()
```

```python
pv.head()
```

```python
pv['체험'] = (pv['스포츠']+pv['체험'])/2
```

```python
pv.head()
```

```python
pv.drop(['숙박','의료','스포츠'], axis = 1, inplace = True)
```
